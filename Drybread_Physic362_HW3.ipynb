```python
#Exercise 1

import numpy as np
import matplotlib.pyplot as plt

weather_data = np.genfromtxt('weather_data.txt',
                            delimiter = ',')

temperature, humidity = weather_data[:, 1], weather_data[:, 2] #temperature and humidity data imported from text file

Max_Temp, Min_Temp = np.max(temperature), np.min(temperature) #using numpy to find min and max temperature

Max_Humid, Min_Humid = np.max(humidity), np.min(humidity)

Average_Temp, Average_Humid = np.mean(temperature), np.mean(humidity) #using numpy to find average temp and humidity

#Timestamp was kept in UTC seconds since Jan 1st 1970, so calculated it to show span of hours
x_values = (weather_data[:, 0] - 1503000000) / (60**2) - 65

temp_y_values, humid_y_values = temperature, humidity

fig, (ax1, ax2) = plt.subplots(2, sharex = True)

#Two Subplots 
ax1.plot(x_values, temp_y_values, 'b--', linewidth=2)
ax1.set_title('Temperature and Humidity vs. Time', fontsize = 14 )
ax1.set_ylabel('Temperature (\N{DEGREE SIGN}C)', fontsize=14)
ax1.set_ylim(Min_Temp, Max_Temp)
ax1.grid()

ax2.plot(x_values, humid_y_values, 'r--', linewidth = 2)
ax2.set_xlabel('Hours')
ax2.set_ylabel('Humidity (%)', fontsize = 14)
ax2.set_ylim(Min_Humid, Max_Humid)
ax2.grid()

plt.savefig("Weather_data.png")

plt.show()

#Extra relevant information
print('Max Temperature', Max_Temp,'\N{DEGREE SIGN}C''\n'
      'Min Temperature', Min_Temp,'\N{DEGREE SIGN}C''\n'
      'Average Temperature', Average_Temp,'\N{DEGREE SIGN}C''\n'
      'Average Humidity', Average_Humid, '%')
```

```python
#Exercise 2

from math import sin, cos

#a) 2(sin(x/2)/x)^2 seems to be better fit for x << 1.
#This is because squaring a very small number makes an
#even smaller number which will cause the first expression
#to go to zero much faster than the second expression. This 
#action would create more errors in the process the computer 
#goes through comparedto the other method.

#b)

for x in [1e-5, 1e-8]:

    Eq1 = ((1 - cos(x)) / (x**2))

    Eq2 = (((sin(x / 2)) / x)**2) * 2

    print('Equation 1 =', '{:.18f}'.format(Eq1))
    print('Equation 2 =', '{:.18f}'.format(Eq2))
```

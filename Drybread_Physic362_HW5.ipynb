```python
#Exercise 1
import numpy as np
from scipy import linalg as LA

#a)
'''
           A     *    v     =   b
     [4 -1 -1 0 0]   [V1]      [5]
     [-1 4 -1 0 0]   [V2]      [5]
     [-1 -1 4 -1 -1] [V_i]     [0]
     [0 0 -1 3 -1]   [V_N-1]   [0]
     [0 0 -1 -1 3]   [V_N]     [0]
'''

#b)
A = np.array(([4, -1, -1, 0, 0], 
              [-1, 4, -1, 0, 0], 
              [-1, -1, 4, -1, -1], 
              [0, 0, -1, 3, -1], 
              [0, 0, -1, -1, 3]))

b = np.array(([5, 5, 0, 0, 0]))

N = 4

for i in range(0, N, 2):
    v = LA.solve(A, b)
print('\nA\n', A, '\nb\n', b, '\nv\n', v)

```

```python
#Exercise 2
import numpy as np
from scipy import linalg as LA
from numpy import linalg as npl
#a) The rank of the matrix is 2 because two of the rows are linearly independent
# This means that the dimension of the nullspace is ...
A = np.array([[2, 4, 6, 8], 
              [-1, 0, 1, 0], 
              [1, 2, 3, 4], 
              [1, 4, 7, 8]])

p, l, u = LA.lu(A)
print('a)\n Permutation Matrix (p)\n', p, 
    '\nLower Diagonal (l)\n', l, 
     '\nUpper Diagonal(u)\n', u)

#b)
eigvals, eigvects = LA.eig(A)
print('Eigenvalues\n', eigvals)

#c) The last two singular values are equal to zero
U, s, Vh = LA.svd(A)
print('Unitary Matrix 1 (U)\n', U,
     '\nSingular Values (s)\n', s,
     '\nUnitary Matrix 2 (Vh)\n', Vh)

#d)
m, n = 4, 4
S = np.zeros((m, n))
for i in range(min(m, n)):
    S[i, i] = s[i]
print('S\n', S)

U = npl.inv(U)
Vh = npl.inv(Vh)

b = np.array([12, 0, 6, 12])


new_A = npl.multi_dot((U, Vh, S))
print('new A\n', new_A)
x, R, r, S_new_A = LA.lstsq(new_A, b)
print('Least Squares Solution\n', x, '\nResidues\n', R, '\nrank\n', r, '\nSingular Values of new A\n', S_new_A)
```

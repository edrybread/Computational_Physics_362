```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# generate fine list of numbers for plotting
x = np.linspace(0, 1, 1000)

# plot the right hand side of the equation
plt.plot(x, (924*x**6 - 2772*x**5 + 3150*x**4 - 1680*x**3 + 420*x**2 -42*x + 1), label='Px')

plt.axhline(0)

# make a nice axis label and legend
plt.xlabel("x")
plt.legend(loc='upper right', fontsize=16, labelspacing=0.3)

# this displays all components of the plot
plt.show()
```

```python
#Exercise 2
#P(x) = 924*x**6 - 2772*x**5 + 3150*x**4 - 1680*x**3 + 420*x**2 -42*x + 1

from numpy import exp, sign
from scipy.misc import derivative as derive
def f(x):
    return 924*x**6 - 2772*x**5 + 3150*x**4 - 1680*x**3 + 420*x**2 -42*x + 1


def deriv(x):
    return derive(f, 1, n = 1)
for i in range(0, 1):
    x = i
    compare_x = 1000
    counter = 0
    accuracy = 1e-10

    while(abs(x-compare_x) > accuracy):
        counter += 1
        compare_x = x
        x = x - f(x)/deriv(x)
    
    print("Newton's method converged to ", x, " after ", counter, " iterations")
```
